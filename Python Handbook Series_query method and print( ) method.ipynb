```python
# 2024/02/22 
import pandas as pd

# Boolean indexing
fooframe = pd.DataFrame({'Size': ['Large', 'Medium', 'Small', 'Tiny'],
                         'Color': [1, 2, 3, 4]})
subframe = fooframe[fooframe['Size'] == 'Large']

# An example of using .query() and chaining
subframe = fooframe.query("Size == 'Large'")

# You can chain this query directly onto the DataFrame construction
subframe = pd.DataFrame({'Size': ['Large', 'Medium', 'Small', 'Tiny'],
                         'Color': [1, 2, 3, 4]}).query("Size == 'Large'")

# Use of @ to reference local variables
my_size = 'Large'
subframe = fooframe.query("Size == @my_size") # This doesn't work for the column labels

# str.format() can be used for the column labels
col_label = 'Size'
my_size = 'Large'
subframe = fooframe.query("{0} == '{1}'".format(col_label, my_size))

# f-strings is pretty
subframe = fooframe.query(f"{col_label} == '{my_size}'")
```

```python
# Print() method source: Print() in Python by Keno Leon in Medium
# Part one: Basics
print('Hello World')

# use \n to separate the output into different lines
print('Hello \nWorld')

# cast numbers as strings
print('this is line: ' + str(1))
print('this is line: ' + str(2))

# Wrap everything in a function and print the output
def gDay():
    return('Good Day Sir !')
print(gDay())

# Wrap everything inside a function
def gDay():
    print('Good Day Madam')
gDay()

# Print using a list, tuple, dictionary, or any iterable
list = [i+1 for i in range(10)]
print(list)
print(list[3])

# To print text with list, tuple, dictionary, or any iterable, add a comma , 
print('LIST:' , list)

# Objects/classes are slightly more complex. 
# Need to differentiate between the class itself and the instance/ instance method
class simpleObject():
    def simpleObject_method(self):
        return ('Hello again')

print(simpleObject) # print the class

a = simpleObject() # create a class instance
print(a) # print the instance
print(a.simpleObject_method())
```

```python
# Part two for print(): Intermediate level
# Concatenating multiple variables along with some text by separating things with +
var_1 = 'This is number: '
numbers = [1, 2, 3]

print( var_1 + str(numbers[0]) + ' The only one !') # interchangable with OUTPUT
print( var_1 + str(numbers[1]))
print( numbers[0] + numbers[1] + numbers[2])

# Python's formatting function:  "string {}".format(value)
print("This is number: {0} the only one !".format(numbers[0]))
print("This is number: {0}".format(numbers[1]))

text = 'Hello there'
print("{}".format(text))

# adding more than one value: "str {} str {} str {}".format(v1, v2, v3)
text_1 = 'Hello Henchman'
text_2 = 'Oddjob'
henchmanRank = [1, 2, 3, 4, 5]

print("{} {} You are my number {}".format(text_1, text_2, henchmanRank[1]))

# Use index to keep track of replacement variables
firstName = 'James'
lastName = 'Bond'

print('The name is  {1}...{0}{1}'.format(firstName, lastName))

# Add a coma :
print('{:,}'.format(1000000))

# Add a coma and display 2 decimal points:
print('${:,.2f}'.format(1000000))

# Express percentage with 2 decimal points
sureness = 80/99
print ('I am sure about this: {:.2%}'.format(sureness))

# Show only 2 decimal places
floaty = 10.2222222
print('floaty value is {:.2f}'.format(floaty))

# Round and show 2 decimals:
floaty = 10.2299999
print('floaty value is {0:.2f}'.format(round(floaty,2)))
# this subject can be extended to dates, tables, hex values and on

# f_strings: f"{value}"
text = 'Hello there'
print(f"{text}")

# f"str {v1} str{v2} str{v3}"
text_1 = 'Hello there'
text_2 = 'General Kenobi'
print(f"{text_1}...{text_2}")

henchmanRank = [1,2,3,4,5]
print(f"{text_1} {text_2} You are my number {henchmanRank[1]}")

# using them with functions
def to_uppercase(text):
    return text.upper()

firstName = 'Bond'
print(f"{to_uppercase(firstName)}!...James {firstName}")

# format numbers using f-strings
# add a coma
print(f"{1000000:,}")

# add a coma and display 2 decimal points
print(f"{1000000:,.2f}")

# express percentages with 2 decimal points
sureness = 80/99
print(f"I am sure about this: {sureness:.2%}")

# show only 2 decimal places
floaty = 10.229999
print(f"floaty value is {floaty:.2f}")

# round and show 2 decimals
print(f"floaty value is {round(floaty,2):.2f}")
```
